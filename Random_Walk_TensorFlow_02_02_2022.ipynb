# 1-Dimensional Random Walk, 1000 Steps

```python
step = 0

rand_walk = [0]

for x in range(1000):
    step = rand_walk[-1]
    walk = np.random.choice([-1, 1])
    step += walk
    rand_walk.append(step)

# print(rand_walk)
plt.plot(rand_walk)
```

## 2-Dimensional Random Walk + Potential Energy Function

```python
# 2-D potential energy function 
# p=ax*x + b*y*y 
# Force in x = dp/dx = -2*x 
# Force in y = dp/dy = -2*y 
# Try different values of a and b, first with a=b, then a  b
```

```python
# Random walk in two dimensions
from pylab import *
dim = 2
N = 10000 # nr of steps
nsample = 1000 # nr of samples
xf = zeros((nsample,dim))
# n = zeros(nsample)
for i in range(nsample):
# Generate random walk of N steps
    z = 2*randint(2,size=(N,dim))-1
    xf[i,:] = sum(z,axis=0)
    # nrm = norm(xf[i])
    # n[i] = nrm

x = [i[0] for i in xf]
y = [i[1] for i in xf]
    

plt.scatter(x, y)
plt.grid()
plt.show()
# #plot(n)
# Nn,edges = histogram(n)
# #3hist(n)
# #%%
# xedges = array(range(int(min(x[:,0])),int(max(x[:,0])),2))+0.5
# yedges = array(range(int(min(x[:,1])),int(max(x[:,1])),2))+0.5
# Nn,xedges,yedges = histogram2d(x[:,0],x[:,1],[xedges,yedges])
# imshow(Nn,interpolation='nearest', origin='low',
# extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]])
# axis('equal')
```

## Random Walk Website Examples

```python
# https://persagen.com/files/misc/random_walks-diffusion.pdf
# Random walk in one dimension - measurement
from pylab import *
N = 100
nsample = int(10000000/N)
x = zeros(nsample)
for i in range(nsample):
    z = 2*np.random.randint(0,2,size=N)-1
    x[i] = sum(z)
edges = array(range(-N,N+1))-0.5
Nx,e = histogram(x,edges)
x = (edges[:-1]+edges[1:])/2.0
dx = diff(edges)
Px = Nx/(dx*nsample)
plot(x,Px), xlabel('x'), ylabel('P(x)')
show()
```

```python
# https://persagen.com/files/misc/random_walks-diffusion.pdf
# Random walk in one dimension - scaling
from pylab import *

Nvalues = [10,50,100,200,400]

for N in [10,50,100,200,400]:
    nsample = int(1000000/N)
    x = zeros(nsample)
    for i in range(nsample):
        z = np.random.randint(3,size=N)-1
        x[i] = sum(z)
    edges = array(range(-N,N+1))-0.5
    Nx,e = histogram(x,edges)
    x = (edges[:-1]+edges[1:])/2.0
    dx = diff(edges)
    Px = Nx/(dx*nsample)
    plot(x,Px)
show()
```

# 2-Dimensional Random Walk

## 1,000 Walkers, 10 Steps

```python
# defining the number of walkers
w = 1000
# defining the number of steps
s = 10
 
#creating two array for containing x and y coordinate
#of size equals to the number of walkers and filled up with 0's
x = np.zeros(w)
y = np.zeros(w)

for walkers in range(w):
    for steps in range(s):
        rand = randint(0,3)
        if rand == 0:
            y[walkers] = y[walkers] + 1
            x[walkers] = x[walkers]
        elif rand == 1:
            x[walkers] = x[walkers] + 1
            y[walkers] = y[walkers]
        elif rand == 2:
            y[walkers] = y[walkers] - 1
            x[walkers] = x[walkers]
        else:
            x[walkers] = x[walkers] - 1
            y[walkers] = y[walkers]
 
plt.scatter(x, y)
plt.grid()
plt.show()
```

```python
def plot_image(i, predictions_array, true_label, img):
  true_label, img = true_label[i], img[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])

  plt.imshow(img, cmap=plt.cm.binary)

  predicted_label = np.argmax(predictions_array)
  if predicted_label == true_label:
    color = 'blue'
  else:
    color = 'red'

  plt.xlabel("{} {:2.0f}% ({})".format(class_names[predicted_label],
                                100*np.max(predictions_array),
                                class_names[true_label]),
                                color=color)

def plot_value_array(i, predictions_array, true_label):
  true_label = true_label[i]
  plt.grid(False)
  plt.xticks(range(10))
  plt.yticks([])
  thisplot = plt.bar(range(10), predictions_array, color="#777777")
  plt.ylim([0, 1])
  predicted_label = np.argmax(predictions_array)

  thisplot[predicted_label].set_color('red')
  thisplot[true_label].set_color('blue')
```

```python
i = 0
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plot_image(i, predictions[i], test_labels, test_images)
plt.subplot(1,2,2)
plot_value_array(i, predictions[i],  test_labels)
plt.show()
```

## Prediction Mistakes Using Confusion Matrix

```python
model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Flatten(input_shape=(28, 28, 1)))

model.add(tf.keras.layers.Dense(units=128,
                       activation='relu'))

model.add(tf.keras.layers.Dense(units=10,
                       activation=None))
```

```python
model.compile(loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              optimizer=tf.keras.optimizers.Adam(),
              metrics=['accuracy'])
```

```python
history =  model.fit(train_images, train_labels, epochs=10)
```

```python
# Evaluate the model
plt.plot(history.history['accuracy'], label='accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
test_labels_pred = []
raw_predictions = model(test_images).numpy()
for row_idx in range(raw_predictions.shape[0]):
    logits = raw_predictions[row_idx, :]
    probs = tf.nn.softmax(logits).numpy()
    class_pred = np.argmax(probs)
    test_labels_pred.append(class_pred)
print(test_labels_pred)
```

```python
print("Classes: \n \
        1. T-shirt/top \n \
        2. Trouser \n \
        3. Pullover \n \
        4. Dress \n \
        5. Coat \n \
        6. Sandal \n \
        7. Shirt \n \
        8. Sneaker \n \
        9. Bag \n \
        10. Ankle boot")
```

```python
mat = confusion_matrix(test_labels, test_labels_pred)
print(mat)
```
